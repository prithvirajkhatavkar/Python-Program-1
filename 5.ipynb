```python
#1 to calculate length of the string
my_string = "Hello"
length = len(my_string)
print(length)


#2 to make the string from 1st and last two characters from given string
text = "Programming"
# Slice first two [0:2] and last two [-2:]
result = text[:2] + text[-2:]
print(result)  # Output: Prng


#3 to concatenate two strings
str1 = "Hello"
str2 = "World"

# Join them with a space
full_string = str1 + " " + str2
print(full_string)  # Output: Hello World
```
