```python

#assignment no.6

#1 Create a list
fruits = ["apple", "banana", "cherry"]

# Check for presence using 'in'
if "apple" in fruits:
    print("Apple is in the list.")

# Check for absence using 'not in'
if "orange" not in fruits:
    print("Orange is not in the list.")


#2 indexing, negative indexing, slicing operations
my_list = [10, 20, 30, 40, 50]

print("Index 0:", my_list[0])         # Positive Indexing
print("Last element:", my_list[-1])   # Negative Indexing


#3 update, Append, insert and delete opertion
data = [1, 2, 3]

# Append
data.append(4)        # [1, 2, 3, 4]

# Insert
data.insert(1, 99)    # [1, 99, 2, 3, 4] (Inserts 99 at index 1)

# Update
data[0] = 100         # [100, 99, 2, 3, 4]

# Delete (using pop or remove)
data.pop(1)           # Removes element at index 1
print("Final list:", data)


#4 basic operations
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Concatenation
combined = list1 + list2      # [1, 2, 3, 4, 5, 6]

# Repetition
repeated = list1 * 2          # [1, 2, 3, 1, 2, 3]

# Length, Max, Min
print("Length:", len(list1))
print("Maximum:", max(list1))
print("Minimum:", min(list1))
```
