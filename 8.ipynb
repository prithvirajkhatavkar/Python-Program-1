```python
# Creating a tuple
t = (1, 2, 3, 4)

# Length
print("Length:", len(t))

# Concatenation
t2 = (5, 6) # Corrected: define t2 for concatenation
print("Concatenation:", t + t2)

# Repetition
print("Repetition:", t * 2)

# Membership
print("Is 3 in tuple?", 3 in t)

t = (10, 20, 30, 40, 50)

# Indexing
print("First element:", t[0])

# Negative indexing
print("Last element:", t[-1])

# Slicing
print("Sliced tuple:", t[1:4])

# Iteration
print("Elements in tuple:")
for i in t:
    print(i)

t = (1, 2, 3)

# Tuples are immutable
# t[0] = 10 # This will give error if uncommented
print("Original tuple:", t)

# Deleting tuple
# del t # Uncommenting and running this would cause a NameError for subsequent print(t)
# print(t) # This will give error because t is deleted

t = (5, 2, 9, 1)

print("Length:", len(t))
print("Maximum:", max(t))
print("Minimum:", min(t))
print("Tuple conversion:", tuple ([1, 2, 3])) # Corrected: added missing '])'
```
